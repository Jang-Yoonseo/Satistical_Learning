```r
crim <- Boston$crim
```

```r
crim1 <- rep(0, length(crim))
crim1[crim >= 0.25] <- 1
```

```r
set.seed(1)
train <- sample(c(TRUE, FALSE), length(crim1), #true,false를 데이터 개수만큼 해서 뽑겠단 의미
    replace = TRUE)
test <- (!train) #훈련평가자료 50:50 분할
```

```r
crim1
```

```r
Boston.train <- Boston[train,]
Boston.test <- Boston[test,]
```

```r
library(MASS)
#LDA 적합
lda.fit <- lda(crim1 ~ lstat + nox + age, data = Boston,
    subset = train)     # LDA 적합
lda.fit
plot(lda.fit)   # 적합결과 시각화 (discriminant function 값)
```

```r
lda.pred <- predict(lda.fit, Boston.test)     # 예측
# 예측성공확률 및 분류결과(class) : 분류결과는 0.5 임계치를 전제로 함
```

```r
lda.class <- lda.pred$class
table(lda.class, Boston.test) #confusion matrix
mean(lda.class == Boston.crim)   # 정분류율
```
